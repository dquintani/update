```python
vac_almenosuna_porc = pd.read_csv("https://raw.githubusercontent.com/dquintani/vacunacion/main/datos/primeras_bidosis_acumulado.csv", index_col=0)
display(vac_almenosuna_porc)

casos_diarios = pd.read_csv("https://github.com/dquintani/covid/raw/main/casos_diarios.csv", index_col=0)
display(casos_diarios)

vac_last_update = vac_almenosuna_porc.index[-1]
covid_last_update = casos_diarios.index[-1]

print("last update vacunacion: ", vac_last_update)
print("last update covid epi: ", covid_last_update)

```

```python
#POSTS DE VACUNACION
####################################################################################################################################################################################
def post_vacunas1(hour):
  if abs(now.hour - hour) < 1:
    api.PostUpdate("#Vacunación en #Bolivia: \nResumen Nacional \nMás en: www.boligrafica.com",
                     media= media_vacunas1)
####################################################################################################################################################################################
    
####################################################################################################################################################################################    
def post_vacunas2(hour):
  if abs(now.hour - hour) < 1:
    api.PostUpdate("#Vacunación en #Bolivia: \nResumen Estadístico \nMás en: www.boligrafica.com", 
                     media= media_vacunas2)
####################################################################################################################################################################################


def post_vacunas3(hour):
  if abs(now.hour - hour) < 1:
    api.PostUpdate("#Vacunación en #Bolivia: \nResumen Detallado",# ({}) \nEn total, hemos aplicado {} dosis a nivel nacional \nMás en www.boligrafica.com".
                  #  format(ayer, int(df1.iloc[-1,2]+df1.iloc[-1,5])), 
                     media= media_vacunas3)

def post_vacunas_duo(hour):
  if abs(now.hour - hour) < 1:
    api.PostUpdate("#Vacunación en #Bolivia: \nResumen Detallado",# ({}) \nEn total, hemos aplicado {} dosis a nivel nacional \nMás en www.boligrafica.com".
                  #  format(ayer, int(df1.iloc[-1,2]+df1.iloc[-1,5])), 
                     media= [media_vacunas1, media_vacunas3])

def post_vacunable_vs_total(hour):
  if abs(now.hour - hour) < 1:
    api.PostUpdate("¡RECORDATORIO SEMANAL!\nMucho cuidado con confundirse entre los porcentajes de vacunación sobre la población *total* y sobre la población *vacunable* (+16 años).", 
                     media= media_vacunable_vs_total)
    


####################################################################################################################################################################################
def post_vacunados_primera_vez(hour):
  if abs(now.hour - hour) < 1:
    api.PostUpdate("#COVID19 en #BOLIVIA: \nCantidad de personas vacunadas contra el COVID por primera vez (solo incluye primeras dosis de esquemas bidosis y Janssen unidosis).",
                     media= media_vacunado_primera_vez)
####################################################################################################################################################################################




#POSTS DE EPIDEMIOLOGIA
####################################################################################################################################################################################
def post_resumen_nacional_epid(hour):
  if abs(now.hour - hour) < 1:
    api.PostUpdate("#COVID19 en #BOLIVIA: \nResumen Nacional (Casos, Muertes, Pruebas y Positividad) (fecha: {}) \nMás en www.boligrafica.com".format(covid_last_update),
                     media= media_resumen_nacional_epid)
####################################################################################################################################################################################



####################################################################################################################################################################################
def post_casos(hour):
  if abs(now.hour - hour) < 1:
    api.PostUpdate("#COVID19 en #BOLIVIA: \nCasos Confirmados ({}) \nMás en: www.boligrafica.com".format(covid_last_update), 
                     media= media_casos)
####################################################################################################################################################################################



####################################################################################################################################################################################
def post_muertes(hour):
  if abs(now.hour - hour) < 1:
    api.PostUpdate("#COVID19 en #BOLIVIA: \nMuertes Diarias ({}) \nMás en: www.boligrafica.com".format(covid_last_update), 
                     media= media_muertes)
####################################################################################################################################################################################


def post_casos_y_muertes(hour):
  if abs(now.hour - hour) < 1:
    api.PostUpdate("#COVID19 en #BOLIVIA: \nCasos Confirmados y Muertes Diarias ({}) \nMás en: www.boligrafica.com".format(covid_last_update), 
                     media= [media_casos, media_muertes])

def post_activos1(hour):
  if abs(now.hour - hour) < 1:
    api.PostUpdate("#COVID19 en #BOLIVIA: \nCasos Activos, resumen estadístico ({}) \nMás en: www.boligrafica.com".format(covid_last_update), 
                     media= media_activos1)

####################################################################################################################################################################################
def post_activos2(hour):
  if abs(now.hour - hour) < 1:
    api.PostUpdate("#COVID19 en #BOLIVIA: \nCasos Activos ({}) \nMás en: www.boligrafica.com".format(covid_last_update), 
                     media= media_activos2)
####################################################################################################################################################################################

def post_activos_duo(hour):
  if abs(now.hour - hour) < 1:
    api.PostUpdate("#COVID19 en #BOLIVIA: \nCasos Activos ({}) \nMás en www.boligrafica.com".format(covid_last_update), 
                     media= [media_activos1, media_activos2])

def post_pruebas1(hour):
  if abs(now.hour - hour) < 1:
    api.PostUpdate("#COVID19 en #BOLIVIA: \nPruebas procesadas de COVID-19, fuente: SEDES departamentales ({}) \n \nMás en www.boligrafica.com".format(covid_last_update), 
                     media= media_pruebas)
  
def post_positividad1(hour):
  if abs(now.hour - hour) < 1:
    api.PostUpdate("#COVID19 en #BOLIVIA: \nPositividad diaria en Bolivia, fuente: SEDES departamentales ({}) \n \nMás en www.boligrafica.com".format(covid_last_update), 
                     media= media_positividad1)
  
def post_positividad2(hour):
  if abs(now.hour - hour) < 1:
    api.PostUpdate("#COVID19 en #BOLIVIA: \nPositividad diaria a nivel nacional y en los 9 departamentos ({}) \n \nMás en www.boligrafica.com".format(covid_last_update), 
                     media= media_positividad2)
def post_positividad_duo(hour):
  if abs(now.hour - hour) < 1:
    api.PostUpdate("#COVID19 en #BOLIVIA: \nPositividad diaria a nivel nacional y en los 9 departamentos ({}) \n \nMás en www.boligrafica.com".format(covid_last_update), 
                     media= [media_positividad1, media_positividad2])
    

####################################################################################################################################################################################
def post_hospital1(hour):
  if abs(now.hour - hour) < 1:
    api.PostUpdate("Hospitalizados con prueba positiva (+) de COVID-19 en Bolivia y los 9 departamentos \n\nFuente: Reportes oficiales del SNIS-ve (contribución anónima)",
                   media = media_hospital)
####################################################################################################################################################################################
```
